```python
# Create a list
numbers = [10, 20, 30, 40, 50]

# Check element presence
print(20 in numbers)      # True
print(60 not in numbers)  # True
```

```python
lst = [1, 2, 3, 4, 5]

# Indexing
print(lst[2])      # 3

# Negative indexing
print(lst[-1])     # 5

# Slicing
print(lst[1:4])    # [2, 3, 4]
```

```python
lst = [10, 20, 30, 40]

# Update
lst[1] = 25

# Append
lst.append(50)

# Insert
lst.insert(2, 15)

# Delete
lst.remove(30)

print(lst)
```

```python
a = [1, 2, 3]
b = [4, 5]

# Concatenation
c = a + b
print(c)

# Repetition
print(a * 2)

# Length
print(len(c))

# Maximum & Minimum
print(max(c))
print(min(c))
```

```python
# Program to declare variables and display their types

integer_value = 10
float_value = 10.5
string_value = "Pragati"
boolean_value = True

print("Integer value:", integer_value, "Type:", type(integer_value))
print("Float value:", float_value, "Type:", type(float_value))
print("String value:", string_value, "Type:", type(string_value))
print("Boolean value:", boolean_value, "Type:", type(boolean_value))
```
