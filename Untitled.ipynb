```python
# Creating a tuple
t1 = (1, 2, 3, 4)

# Finding length
print("Length of tuple:", len(t1))

# Concatenation
t2 = (5, 6)
t3 = t1 + t2
print("After concatenation:", t3)

# Repetition
print("After repetition:", t1 * 2)

# Membership
print("Is 3 present in tuple?", 3 in t1)
```
