```python
# Demonstration of different Python Data Types

# Integer
integer_value = 25
print("Integer value:", integer_value)
print("Type:", type(integer_value))
print()

# Float
float_value = 10.5
print("Float value:", float_value)
print("Type:", type(float_value))
print()

# String
name = "Pragati"
print("String value:", name)
print("Type:", type(name))
print()

# Boolean
is_student = True
print("Boolean value:", is_student)
print("Type:", type(is_student))
print()

# List
fruits = ["Apple", "Banana", "Mango"]
print("List value:", fruits)
print("Type:", type(fruits))
print()

# Tuple
numbers = (1, 2, 3, 4)
print("Tuple value:", numbers)
print("Type:", type(numbers))
print()

# Dictionary
student = {"Name": "Pragati", "Age": 20, "Course": "B.Tech"}
print("Dictionary value:", student)
print("Type:", type(student))
print()

# Set
set_values = {10, 20, 30, 40}
print("Set value:", set_values)
print("Type:", type(set_values))
```
